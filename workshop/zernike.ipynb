Zernike Modes
============

In this notebook we show how to use aotools to generate arrays of Zernike modes.

```python
import aotools
import numpy
import matplotlib.pylab as plt
```

The simplest way to generate Zernike modes is to use the `aotools.ZernikeArray` function. this generates all of the Zernike modes up to a certain J index. By default the amplitude of the Zernikes will be scaled to the Noll normalisation, however, this can be changed to peak to valley (`"p2v"`) or route mean square (`"rms"`) normalisation by specifying the `norm` kwarg.

```python
zernike_array = aotools.zernikeArray(10, 128)
```

```python
# Plot the first few Zernike modes created
fig, ax = plt.subplots(nrows=3, ncols=3)
counter=0
for i in range(3):
    for j in range(3):
        counter+=1
        im = ax[i, j].imshow(zernike_array[counter], interpolation='nearest', vmin=zernike_array.min(), vmax=zernike_array.max())
        ax[i, j].axis('off')
```

Alternatively, if only a single Zernike mode is required then an array containing a single mode can be generated using the `aotools.zernike_nm` function where the [n,m] indeces are required, or the `aotools.zernike_noll` function where the J index is specified.

```python
zernike_array = aotools.zernike_nm(3, 1, 128)

plt.figure()
plt.imshow(zernike_array)
```

```python
zernike_array = aotools.zernike_noll(8, 128)

plt.figure()
plt.imshow(zernike_array)
```

There are a number of other functions available for manipulating Zernike modes, including:
- converting J index to the [n,m] convention (`aotools.ZernIndex`)
- calculation of Zernike radial functions (`aotools.PhaseFromZernikes`)
- Creating a single array which is a sum of multiple Zernike modes (`aotools.zernikeRadialFunc`)
